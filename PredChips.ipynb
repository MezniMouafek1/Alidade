```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
import lightgbm as lgb

#LightGBMRegressor
gbm0 = lgb.LGBMRegressor(
    objective='regression',
    num_leaves=11,
    learning_rate=0.006,
    n_estimators=2000)
```

```python
gbm0.fit(X_train, y_train, eval_metric='rmse')
```

```python
X_test_pred_light
```

```python
import xgboost as xgb

reg = xgb.XGBRegressor(n_estimators=2000,
                       nthread=4,
                       min_child_weight=6,
                       learning_rate=0.01,
                       subsample=0.7,
                       max_depth=6,
                       colsample_bytree=0.6)
reg.fit(X_train, y_train,
        eval_set=[(X_test, y_test)],
        early_stopping_rounds=25, #stop if n consequent rounds without decrease of error
        verbose=250) # Change verbose to True if you want to see it train
X_test_pred = reg.predict(X_test)
# rmse = root mean squared error
# racine moyenne carré de l'erreur 
```

```python
help(xgb.XGBRegressor())
```

```python
X_test_pred
```

```python
from xgboost import plot_importance
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=(10 ,10))
plot_importance(reg, ax=ax)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
#Visualizing Test Results 
plt.scatter(X_test.index,y_test, color= 'red')
plt.scatter(X_test.index,predictions,color='green')
plt.title ("Visuals Linear regression")
plt.xlabel("id")
plt.ylabel("visibilité")
plt.show()
```

```python
#Visualizing XGBOOST Test Results 
plt.scatter(X_test.index,y_test, color= 'red')
plt.scatter(X_test.index,X_test_pred,color='blue')
plt.title ("Visuals for XGBoost test")
plt.xlabel("id")
plt.ylabel("visibilité")
plt.show()
```

```python
#Visualizing Light gbm Test Results 
plt.scatter(X_test.index,y_test, color= 'red')
plt.scatter(X_test.index,X_test_pred_light,color='yellow')
plt.title ("Visuals for Light gbm test")
plt.xlabel("id")
plt.ylabel("visibilité")
plt.show()
```

```python
from catboost import Pool, CatBoostRegressor

#CatBoostRegressor
train_pool = Pool(X_train, y_train) 
test_pool = Pool(X_test)
cbm0 = CatBoostRegressor(depth=10, learning_rate=0.01, eval_metric='RMSE')
cbm0.fit(train_pool)

X_test_pred_cat = cbm0.predict(test_pool)
```

```python
#Visualizing Light gbm Test Results 
plt.scatter(X_test.index,y_test, color= 'red')
plt.scatter(X_test.index,X_test_pred_cat,color='black')
plt.title ("Visuals for catboost test")
plt.xlabel("id")
plt.ylabel("visibilité")
plt.show()
```
